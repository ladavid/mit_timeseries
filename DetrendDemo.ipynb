# Detrending auto-correlated data

A demonstration of spurious cross-correlations originating from auto-correlated time-series, and how detrending can resolve. Our example revolves around two random walks, which one might not think to be cross-correlated. However, any two random walks (if long enough) are almost always correlated with extremely low p-values. This is due in part to random walks' autocorrelation, meaning that their position at any time is highly dependent on their position at the previous time step. Below, we show 1) how two random walks exhibit strong cross-correlations, as well as individual auto-correlations. We then show 2) how removing trends from auto-correlated time-series, and instead performing cross-correlations on residual time-series, leads to non-spurious correlations.

## 1. Generate two random walks. 

Intuititively, one would not think that two random walks should show a temporal correlation. But, due to auto-correlative effects, they do!

```python
import pylab as pl
import scipy.stats as ss
```

```python
# simulate two random walks using a series of random coinflips. the random walk is the cumulative sum of the number of heads (+1)
# or tails (-1) flipped. 
coinflip1_v = sign(1*(np.random.random(1000) > 0.5)-0.5)
coinflip2_v = sign(1*(np.random.random(1000) > 0.5)-0.5)
x1_v = cumsum(coinflip1_v)
x2_v = cumsum(coinflip2_v)
```

```python
# visualize random walks. usually, these don't look cross-correlated.
pl.figure()
pl.plot(x1_v)
pl.plot(x2_v,'r')
pl.grid()
```

But, random walks are highly auto-correlated.

```python
pl.figure()
pl.plot(x1_v[:-1],x1_v[1:],'o')
pl.figure()
pl.plot(x2_v[:-1],x2_v[1:],'or')
```

```python
print "Random walk 1 auto-correlaton (rho): %1.2f" % ss.pearsonr(x1_v[:-1],x1_v[1:])[0]
print "Random walk 2 auto-correlaton (rho): %1.2f" % ss.pearsonr(x2_v[:-1],x2_v[1:])[0]
```

And, their cross correlations will often have extremely low p-values

```python
xcorr_t = ss.pearsonr(x1_v,x2_v)
print "Random walk cross-correlation (rho): %1.2f" % xcorr_t[0]
print "Random walk cross-correlation (pval): %1.4e" % xcorr_t[1]
```

## 2. Detrending the random walks.

A common strategy for resolving the apparent cross-correlation between auto-correlated time-series, is to remove the trends within each time-series and instead study residual time-series. If the trends are correctly inferred, the residuals should not be autocorrelated. Below, we use ARIMA models (a common time-series tool) to detrend our random walks. 

```python
# this example calls R code. to run it, you'll need to install R, the rpy2 Python module, and the 'forecast' R package.
import rpy2.robjects as R
import rpy2.robjects.numpy2ri
rpy2.robjects.numpy2ri.activate()
R.r('library(forecast)');
```

```python
R.r.assign('x1_v',x1_v)
R.r.assign('x2_v',x2_v)

R.r('x1_arima<-auto.arima(x=x1_v,max.p=2,d=1,max.q=2,ic="bic",approximation=FALSE)');
R.r('x2_arima<-auto.arima(x=x2_v,max.p=2,d=1,max.q=2,ic="bic",approximation=FALSE)');

residual1_v = np.array(R.r('x1_arima$residuals'))
residual2_v = np.array(R.r('x2_arima$residuals'))
```

The residual time-series resemble the series of coinflips that gave rise to the random walk. 

```python
pl.plot(residual1_v)
pl.ylim([-1.5,1.5])
```

```python
pl.plot(residual2_v,'r')
pl.ylim([-1.5,1.5])
```

The residual time-series should not be autocorrelated

```python
print "Residual 1 auto-correlaton (rho): %1.3f" % ss.pearsonr(residual1_v[:-1],residual1_v[1:])[0]
print "Resdiual 2 auto-correlaton (rho): %1.3f" % ss.pearsonr(residual2_v[:-1],residual2_v[1:])[0]
```

And, the cross-correlation of the random walks' residuals should not be correlated. 

```python
xcorr_t = ss.pearsonr(residual1_v,residual2_v)
print "Random walk cross-correlation (rho): %1.2f" % xcorr_t[0]
print "Random walk cross-correlation (pval): %1.2f" % xcorr_t[1]
```
